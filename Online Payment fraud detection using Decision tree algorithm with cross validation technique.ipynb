```python
x=data1.iloc[:,:-1].values
y=data1.iloc[:,-1].values
```

### Standardized

```python
from sklearn.preprocessing import StandardScaler
s1=StandardScaler()
s2=s1.fit_transform(x)
s2
```

```python
data2=pd.DataFrame(s2)
data2
```

```python
x1=x
y1=y
```

```python
from imblearn.over_sampling import SMOTE
sm=SMOTE()
x_data,y_data=sm.fit_resample(x1,y1)
```

```python
from collections import Counter
print(Counter(y_data))
```

### Train_test_split

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x_data,y_data,test_size=0.2,random_state=0)
```

### Using Decision Tree

### What is decision tree?
#### Decision Tree is classification supervised machine learning algorithm.
#### The gole is to create model that predicts the value of target variable by learning simple decision rule from
#### data feacture.
#### It learn from the conditional concepts like decision rule.
#### Every decision rules has some conditions.
#### The dipper the tree more complex the decision rule and fitter the model.

```python
from sklearn import tree
d1=tree.DecisionTreeClassifier()
d1.fit(x_train,y_train)
```

```python
y_pred=d1.predict(x_test)
y_pred
```

```python
from sklearn.metrics import accuracy_score
acc=accuracy_score(y_pred,y_test)*100
print(acc)
```

### With Corss-validation technique

```python
from sklearn.model_selection import KFold
k1=KFold(n_splits=5,random_state=11,shuffle=True)
k1.get_n_splits(x1)
print(k1)
```

```python
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import cross_val_predict
score=cross_val_score(d1,x_train,y_train,cv=k1)*100
pred=cross_val_predict(d1,x_test,y_test)
print(np.mean(score))
print(pred)
```
